## Filter Dataset 

```python
df.dropna(subset=["Website"], inplace=True)
df = df[df["Industries"].str.contains('Apparel & Fashion')]
df = df[df["Locations"].str.contains('Germany')]
df.head()
```

#### Filter for Shopify shops 

```python
target_company_forms:list = ["Public Company", "Nonprofit", "Educational"]
df = df[df["Company type"].isin(target_company_forms)==False]
```

```python
df.shape[0]
```

```python
df['shopify'] = df.apply(lambda row: filter_for_shopify(f"http://{row['Domain']}"), axis=1)
```
